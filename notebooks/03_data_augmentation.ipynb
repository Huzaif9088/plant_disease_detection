```python
import tensorflow as tf

import matplotlib.pyplot as plt

from tensorflow.keras import layers
```

```python
data_augmentation = tf.keras.Sequential([
    
    layers.RandomFlip("horizontal"),

    layers.RandomRotation(0.2),

    layers.RandomZoom(0.2),

    layers.RandomContrast(0.2)

])
```

```python
from PIL import Image

import numpy as np

image_path = "/workspaces/plant_disease_detection/data/raw/Tomato_Leaf_Mold/fd686c68-9a4a-4d11-98fa-7c612aff9fd4___Crnl_L.Mold 6974.JPG"

img = Image.open(image_path)

img = img.resize((224,224))

img_array = np.array(img)
```

```python
print(img_array.shape)
```

```python
img_array = tf.expand_dims(img_array, axis=0)
```

```python
print(img_array.shape)
```

```python
plt.figure(figsize=(12,12))

for i in range(9):

    augmented_image = data_augmentation(img_array)

    ax = plt.subplot(3,3,i+1)

    plt.imshow(
        augmented_image[0].numpy().astype("uint8")
    )

    plt.axis("off")

plt.show()
```
